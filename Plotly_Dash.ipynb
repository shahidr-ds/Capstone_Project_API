```python
from jupyter_dash import JupyterDash
from dash import dcc, html
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from dash.dependencies import Input, Output

# Create synthetic data mimicking your variables
np.random.seed(42)
dates = pd.date_range(start='1980-01-01', end='2013-12-31', freq='M')
n = len(dates)
vehicle_types = ['Supperminicar', 'Smallfamiliycar', 'Mediumfamilycar', 'Executivecar', 'Sports']

df = pd.DataFrame({
    'Date': dates,
    'Recession': np.random.choice([0, 1], size=n, p=[0.8, 0.2]),
    'Automobile_Sales': np.random.randint(50, 500, size=n),
    'GDP': np.random.uniform(20000, 50000, size=n),
    'unemployment_rate': np.random.uniform(3, 12, size=n),
    'Consumer_Confidence': np.random.uniform(50, 100, size=n),
    'Seasonality_Weight': np.random.uniform(0.5, 1.5, size=n),
    'Price': np.random.uniform(15000, 40000, size=n),
    'Advertising_Expenditure': np.random.uniform(1000, 10000, size=n),
    'Vehicle_Type': np.random.choice(vehicle_types, size=n),
    'Competition': np.random.uniform(0, 1, size=n)
})
df['Year'] = df['Date'].dt.year

recession_df = df[df['Recession'] == 1]

app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("XYZAutomotives Dashboard - Recession Report", style={'textAlign': 'center'}),
    
    # Graph 1: Avg Sales fluctuation over recession years (line)
    dcc.Graph(
        id='avg-sales-fluctuation',
        figure=px.line(
            recession_df.groupby('Year')['Automobile_Sales'].mean().reset_index(),
            x='Year',
            y='Automobile_Sales',
            title='Average Automobile Sales Fluctuation Over Recession Period (Year-wise)'
        )
    ),
    
    # Graph 2: Avg vehicles sold by Vehicle Type (bar)
    dcc.Graph(
        id='avg-vehicles-by-type',
        figure=px.bar(
            recession_df.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index(),
            x='Vehicle_Type',
            y='Automobile_Sales',
            title='Average Number of Vehicles Sold by Vehicle Type During Recession'
        )
    ),
    
    # Graph 3: Total advertisement expenditure share by Vehicle Type (pie)
    dcc.Graph(
        id='ad-exp-share',
        figure=px.pie(
            recession_df.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index(),
            names='Vehicle_Type',
            values='Advertising_Expenditure',
            title='Total Advertisement Expenditure Share by Vehicle Type During Recession'
        )
    ),
    
    # Graph 4: Effect of Unemployment Rate on Vehicle Type and Sales (grouped bar)
    dcc.Graph(
        id='unemployment-effect',
        figure=go.Figure(
            data=[
                go.Bar(
                    name='Average Automobile Sales',
                    x=recession_df.groupby('Vehicle_Type')['Automobile_Sales'].mean().index,
                    y=recession_df.groupby('Vehicle_Type')['Automobile_Sales'].mean().values
                ),
                go.Bar(
                    name='Average Unemployment Rate',
                    x=recession_df.groupby('Vehicle_Type')['unemployment_rate'].mean().index,
                    y=recession_df.groupby('Vehicle_Type')['unemployment_rate'].mean().values
                )
            ],
            layout=go.Layout(
                title='Effect of Unemployment Rate on Vehicle Type and Sales During Recession',
                barmode='group',
                yaxis_title='Value'
            )
        )
    )
])

app.run_server(mode='inline')
```

```python
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import pandas as pd
import numpy as np
import plotly.express as px

# Create synthetic data (1980-2013 monthly data)
np.random.seed(0)
dates = pd.date_range(start='1980-01-01', end='2013-12-31', freq='M')
vehicle_types = ['Supperminicar', 'Smallfamiliycar', 'Mediumfamilycar', 'Executivecar', 'Sports']

df = pd.DataFrame({
    'Date': dates,
    'Year': dates.year,
    'Month': dates.month,
    'Automobile_Sales': np.random.randint(50, 500, size=len(dates)),
    'Advertising_Expenditure': np.random.uniform(1000, 10000, size=len(dates)),
    'Vehicle_Type': np.random.choice(vehicle_types, size=len(dates))
})

# Group for yearly average sales (whole period)
yearly_avg_sales = df.groupby('Year')['Automobile_Sales'].mean().reset_index()

app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("XYZAutomotives Dashboard - Yearly Report Statistics",
            style={'textAlign': 'center', 'marginBottom': '40px'}),
    
    html.Div([
        html.Label("Select Year:"),
        dcc.Dropdown(
            id='year',
            options=[{'label': str(y), 'value': y} for y in sorted(df['Year'].unique())],
            value=1980,
            style={'width': '200px'}
        )
    ], style={'marginBottom': '30px', 'textAlign': 'center'}),
    
    # Graph 1: Yearly Automobile Sales (Line chart)
    html.Div([
        dcc.Graph(
            id='yearly-sales-line',
            figure=px.line(yearly_avg_sales, x='Year', y='Automobile_Sales',
                           title='Yearly Average Automobile Sales (1980-2013)')
        ),
        html.P("This line chart shows the average automobile sales for each year across the entire period.",
               style={'textAlign': 'center', 'fontStyle': 'italic', 'marginBottom': '40px'})
    ]),
    
    # Graph 2: Total Monthly Automobile Sales for selected year (Line chart)
    html.Div([
        dcc.Graph(id='monthly-sales-line'),
        html.P("This line chart displays the total automobile sales for each month of the selected year.",
               style={'textAlign': 'center', 'fontStyle': 'italic', 'marginBottom': '40px'})
    ]),
    
    # Graph 3: Average Vehicles Sold by Vehicle Type in selected year (Bar chart)
    html.Div([
        dcc.Graph(id='avg-vehicles-bar'),
        html.P("This bar chart shows the average number of vehicles sold by vehicle type in the selected year.",
               style={'textAlign': 'center', 'fontStyle': 'italic', 'marginBottom': '40px'})
    ]),
    
    # Graph 4: Total Advertisement Expenditure for each vehicle type in selected year (Pie chart)
    html.Div([
        dcc.Graph(id='ad-exp-pie'),
        html.P("This pie chart represents the total advertising expenditure for each vehicle type in the selected year.",
               style={'textAlign': 'center', 'fontStyle': 'italic', 'marginBottom': '40px'})
    ]),
])

@app.callback(
    [Output('monthly-sales-line', 'figure'),
     Output('avg-vehicles-bar', 'figure'),
     Output('ad-exp-pie', 'figure')],
    [Input('year', 'value')]
)
def update_yearly_reports(selected_year):
    # Filter data for selected year
    year_data = df[df['Year'] == selected_year]
    
    # Monthly total sales
    monthly_sales = year_data.groupby('Month')['Automobile_Sales'].sum().reset_index()
    fig_monthly_sales = px.line(monthly_sales, x='Month', y='Automobile_Sales',
                               title=f'Total Monthly Automobile Sales in {selected_year}',
                               markers=True)
    fig_monthly_sales.update_layout(xaxis=dict(tickmode='linear'))
    
    # Average vehicles sold by vehicle type
    avg_vehicles = year_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
    fig_avg_vehicles = px.bar(avg_vehicles, x='Vehicle_Type', y='Automobile_Sales',
                             title=f'Average Vehicles Sold by Vehicle Type in {selected_year}')
    
    # Total advertisement expenditure by vehicle type
    ad_exp = year_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index()
    fig_ad_exp = px.pie(ad_exp, names='Vehicle_Type', values='Advertising_Expenditure',
                       title=f'Total Advertisement Expenditure by Vehicle Type in {selected_year}')
    
    return fig_monthly_sales, fig_avg_vehicles, fig_ad_exp

app.run_server(mode='inline')
```
